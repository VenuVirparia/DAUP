```python
import pandas as pd
import numpy as np
import sklearn.datasets as data
import matplotlib.pyplot as plt

d1 = data.load_iris()
d1 

iris=pd.DataFrame(data=d1.data, columns=d1.feature_names)
iris

iris['target']=d1.target

X = d1.data
y = d1.target


# y_test.size
```

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=3)
```

```python
kmean.fit(X)
```

```python
y_lab = kmean.predict(X)
# y_lab.size
```

```python
result = pd.DataFrame({"Actual": y, "Predicted":y_lab})
result
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print(accuracy_score(y,y_lab))
print(confusion_matrix(y,y_lab))
print(classification_report(y,y_lab))
```

## DBScan

```python
from sklearn.cluster import dbscan
dbs, db2 = dbscan(X)
```

```python
print(accuracy_score(y,db2))
```
