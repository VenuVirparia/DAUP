## KMeans and DBScan

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y_true = make_blobs(n_samples=100,centers=3, cluster_std=0.6) #standard deviation 60% , if similarity>=60% then take it in cluster
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
#km
```

```python
km.fit(X)
y_lab = km.predict(X)
#y_lab
```

```python
plt.scatter(X[:,0], X[:,1],c=y_lab, s=50,cmap='plasma') #viridis
centers = km.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black', s=200,alpha=0.7) #s=size, alpha=transperency/opacity
```
